```python
#To find pokemon acc to numero
def find_poke(x):
        chosen = pokedex.loc[pokedex['NUMERO'] == x ]
        #chosen = chosen.drop(columns = ["NUMERO", "NOM","TYPE_1","TYPE_2","LEGENDAIRE"])     
        return chosen
find_poke(1)
```

```python
import pickle
with open("model/model_forest", "rb") as f:
    model = pickle.load(f)
```

```python
test_df = pd.read_csv("datas/tests.csv")
test_df
```

```python
rand_row = test_df.sample()
rand_row
```

```python
f_poke = find_poke(rand_row["First_pokemon"].iloc[0])
f_poke
```

```python
s_poke = find_poke(rand_row["Second_pokemon"].iloc[0])
a = s_poke.NOM
print(a)
```

```python
def whowillwin(f,s,pokedex):
    first = find_poke(f)
    second = find_poke(s)
    print("First Pokemon is:")
    display(first)
    print("Second Pokemon is:")
    display(second)
    
    a = first.NOM
    b = second.NOM
    
    first_dropped = first.drop(columns = ["NUMERO", "NOM", "TYPE_1", "TYPE_2", "LEGENDAIRE"])
    second_dropped = second.drop(columns = ["NUMERO", "NOM", "TYPE_1", "TYPE_2", "LEGENDAIRE"])
    
    f_pred = model.predict(first_dropped)
    s_pred = model.predict(second_dropped)
    
    if (f_pred[0] > s_pred[0]):
        print(a," is the winner !!!")
    else:
        print(b,"Pokemon 2 is the winner !!!")
```

```python
whowillwin(rand_row["First_pokemon"].iloc[0],rand_row["Second_pokemon"].iloc[0],pokedex)
```
